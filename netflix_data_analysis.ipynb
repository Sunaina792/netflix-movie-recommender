# data preprocessing

```python
df['Release_Date'] = pd.to_datetime(df['Release_Date'])
print(df['Release_Date'].dtype)
```

```python
df['tags'] = df['Overview'] + df['Genre']
df['content'] = df['tags'] + df['Original_Language']
```

```python
import string

def clean_text(text):
    text = str(text).lower()                     # Lowercase
    text = text.translate(str.maketrans('', '', string.punctuation))  # Remove punctuation
    return text

df['Overview'] = df['Overview'].apply(clean_text)
df['Genre'] = df['Genre'].apply(clean_text)
df['Original_Language'] = df['Original_Language'].apply(clean_text)
```

## model building

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer(stop_words='english')
```

```python
tfidf_matrix = tfidf.fit_transform(df['content'])
```

```python
from sklearn.metrics.pairwise import cosine_similarity

cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)
```

```python
indices = pd.Series(df.index, index=df['Title']).drop_duplicates()
```

```python
# df['unique_id'] = df['Title'] + " (" + df['Release_Date'].astype(str) + ")"
# indices = pd.Series(df2.index, index=df['unique_id']).drop_duplicates()
```

```python
def recommend(title, num=5):
    if title not in indices:
        return f"No movie found with title '{title}'"
    
    idx = indices[title]
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)[1:num+1]
    movie_indices = [i[0] for i in sim_scores]
    
    return df[['Title', 'Release_Date']].iloc[movie_indices]
```

```python
recommend("The Batman", 5)
```

```python
def recommend_by_genre(genre, num=5):
    genre = genre.lower()
    # Filter rows that contain the genre
    genre_df = df[df['content'].str.lower().str.contains(genre)]

    # Vectorize the content of these filtered movies
    tfidf = TfidfVectorizer(stop_words='english')
    tfidf_matrix = tfidf.fit_transform(genre_df['content'])

    # Compute similarity within this filtered group
    cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)

    # Recommend from the top N similar to the first one
    sim_scores = list(enumerate(cosine_sim[0]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)[1:num+1]
    movie_indices = [i[0] for i in sim_scores]

    return genre_df[['Title', 'content']].iloc[movie_indices]
```

```python
recommend_by_genre("Crime", 5)
```

```python
def recommend_by_cast(actor_name, num=5):
    actor_name = actor_name.lower()

    # Filter movies where actor is mentioned in the content
    actor_df = df[df['content'].str.lower().str.contains(actor_name)]

    if actor_df.empty:
        return f"No movies found with actor '{actor_name}'"

    # TF-IDF only on filtered data
    tfidf = TfidfVectorizer(stop_words='english')
    tfidf_matrix = tfidf.fit_transform(actor_df['content'])

    cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)

    # Recommend similar to the first movie in the list
    sim_scores = list(enumerate(cosine_sim[0]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)[1:num+1]
    movie_indices = [i[0] for i in sim_scores]

    return actor_df[['Title', 'content']].iloc[movie_indices]
```

```python
recommend_by_cast("Peter Parker", 5)
```

```python
df['content'] = df['content'].str.lower()

```

```python
def recommend_by_language(language, num=5):
    language = language.lower()

    # Filter movies by language
    lang_df = df[df['content'].str.contains(language)]

    if lang_df.empty:
        return f"No movies found in language '{language}'"

    # TF-IDF on filtered language-based content
    tfidf = TfidfVectorizer(stop_words='english')
    tfidf_matrix = tfidf.fit_transform(lang_df['content'])

    cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)

    # Recommend similar to the first one
    sim_scores = list(enumerate(cosine_sim[0]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)[1:num+1]
    movie_indices = [i[0] for i in sim_scores]

    return lang_df[['Title', 'content']].iloc[movie_indices]
```

```python
recommend_by_language("en", 5)
```
